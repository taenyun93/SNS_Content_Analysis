```python
#video와 photo로 타입 분리
def video_photo_binarizer(x):
    list_video = ['VIDEO','video','동영상']
    list_photo = ['PICTURE','picture','photo','GIF']
    
    if x in list_video:
        result = 1
    elif x in list_photo:
        result = 0
    else:
        result = np.nan
    return result
    
    
df_total_clean['type_video'] = df_total_clean.Type.apply(lambda x : video_photo_binarizer(x))
```
